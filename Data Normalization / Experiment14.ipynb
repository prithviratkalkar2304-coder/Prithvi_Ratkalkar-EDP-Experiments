#Data Normalization and Data Type Conversion(Part A-Data Normalization)

```python
#DataSet
import pandas as pd
import numpy as np
data={
    'Product':['Laptop','Mobile','Tablet','Headphones','SmartWatch','Camera'],
    'Price':[55000,20000,30000,45000,5000,120000],
    'Units_Sold':[25,60,23,45,65,8],
    'Discount':[10,20,40,35,16,30]
}
df=pd.DataFrame(data)
print("Original Data:")
print(df)
```

```python
#1 min-max normalization
#min-max normalization for price
df['Price_MinMax']=(df['Price']-df['Price'].min())/(df['Price'].max()-df['Price'].min())
print("\nMin-Max Normalized Price:")
print(df[['Product','Price','Price_MinMax']])
```

```python
#Normalize Multiple Columns
cols=['Price','Units_Sold','Discount']
df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
print("\nNormalized Multiple Columns:")
print(df_norm)
```

```python
#2 Z-score normalization
df['Units_Zscore']=(df['Units_Sold']-df['Units_Sold'].mean())/df['Units_Sold'].std()
print("\nZ-score Normalized Units_Sold:")
print(df[['Product','Units_Sold','Units_Zscore']])
```

```python
#3 Decimal Scaling
df['Price_Decimal']=df['Price']/1000000
print("\nDecimal Scaled Price:")
print(df[['Product','Price','Price_Decimal']])
```

```python
#DataSet
import pandas as pd
import numpy as np
data={
    'Order_ID':[101,102,103,104,105,106],
    'Customer_Gender':['Male','Female','Male','Female','Male','Female'],
    'Payment_Method':['UPI','Credit Card','UPI','Cash on Delivery','UPI','Credit Card'],
    'Product_Category':['Electronics','Clothing','Home','Electronics','Beauty','Electronics'],
    'City':['Pune','Mumbai','Delhi','Bangalore','Pune','Hyderabad'],
    'Order_Value':[4000,2300,3200,5200,1000,4300]
}
df=pd.DataFrame(data)
print("Original Data:")
print(df)
```

```python
#method 1 Label Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender_Label']=le.fit_transform(df['Customer_Gender'])
df
```

```python
#method 2 one hot coding
df_encoded=pd.get_dummies(df,columns=['Payment_Method'])
df_encoded
```

```python
#Method3 Encoded multiple columns
df_multi=pd.get_dummies(df,columns=['Product_Category','City'])
df_multi
```
